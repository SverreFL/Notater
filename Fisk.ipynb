# Fiskeridirektoratet

Fiskeridirektoratet er myndighetenes rådgivende og utøvende organ innen fiskeri- og havbruksforvaltning i Norge. Det er underlagt Nærings- og fiskeridepartementet. Det overordende målet er å sikre rammevilkårene for en lønsomm og bærekraftig fiskeri- og havbruksnæring.

Det har hovedkontor i Bergen og seks regionskontor, samt en del lokale fiskerikontor.

# Risikostyring 

Risikostyring er et etablert begrep i regulering av næringer. Det har blant annet blitt anvendt i reguleringen av Oljenæringen der økonomisk gevinst av å åpne oljefelt må veies opp mot mulige negative miljøkonsekvenser.

Generelt har reguleringen en overordnet målsetning. Størrelsen på risikoen avhenger av sannsynligheten for at en hendelse som påvirker måloppnåelsen negativt inntreffer, samt konsekvense av dette. Denne risikoen må veies opp mot forventede gevinster.

For å håndtere risiko på en systematisk måte må vi kvantifisere både sannsynlighet og konsekvenser. 

Det er nært beslektet med kostnad-nytte analyser, men mer fokus på risikobegrepet ... Kan for eksempel bruke føre-var prinsipp dersom det er vanskelig å beregne risiko. 

## Målstyring

Risikostyring har et klart definert mål. Kan deles opp i delmål og forsøke å kvantifisere måloppnåelse. Kan dele inn i to hoveddeler
1. Identifisere risikoer
    - Kvantifisere omfang/sannsynlighet av hendelser som påvirker måloppnåelse negativt og konsekvenser av dette
    - Identifisere hvordan det kan oppdages/håndheves 
2. Håndheve
    - Med utgangspunkt i analyse over kan det implementers tiltak for å begrense risiko som har blitt identifisert
    - Hvilke tiltak som er aktuelle kan avhenge av alvorlighetsgrad og årsak til brudd på regulering (tilsyn, veiledning, staff)

### Fiskedumping

Med noen unntakt skal all fisk som blir fanget føres til land og registreres. Dette innebærer at det ikke er lov til å dumpe fisk tilbake i havet. Slik dumping medfører sløseri med ressurser samt at vi får dårligere informasjonsgrunnlag om det reelle uttaket. Dette er i konflikt med målsetningen om bærekraftig forvaltning og er ulovlig.

Det kan likevel være økonomiske incentiv til å dumpe fisk (feil type, for store, for små ??). For å sikre at reglene blir fulgt må de håndheves. Dette innebærer at vi må avdekke fartøy som dumper fisk og påføre sanksjoner. Jeg tror det kan være vanskelig å oppdage fordi vi nesten må ta de på fersken, og det er begrensede ressurser til overvåking av flåten.

Vi kan forsøke å bruke data til å gjøre overvåkingen mer målrettet mot fartøy der vi anslår at det er større sannsynlighet for at det dumpes fisk. Aktuelle data er fra sluttseddelen og journal/gps... En indikasjon på fiskedumping kan for eksempel være at fangstsammensettingen (arter/vekt) avviker fra fangsten hos sammenlignbare fartøy (samme redskap i samme område til omtrent samme tid).

## Litt generell teori om regulering

Vi kan analysere dette innenfor rammeverket til økonomisk teori med rasjonelle, optimaliserende aktører.

Reguleringsmyndighetene har et betinget optimeringsproblem der de med begrensede ressurser forsøker å oppnå mål fastlagt av politiske myndigheter. De har begrenset informasjon og kan ikke direkte styre valgene til aktører som blir regulert. For at reglene skal bli overholdt er det nødvendig at aktørene ser at det er i sin egeninteresse å følge reglene. I en enkel modell avhenger av sannsynlighet for å blir tatt og størrelsen på straff. I tillegg kan vi se på litt andre forklaringer
1. Kostnad ved å sette seg inn i regler. Bør være enkle å forstå og overholde.
2. Tillit til reguleringsmyndighet. Både at like og rettferdige vilkår for ulike aktører, reguleringer som er godt begrunnet og som bli overholdt. Større sannsynlighet for å bryte reglene dersom man føler at de er urimelige eller at andre jukser og får fordel av dette

### Informasjon

Informasjon er et sentral aspekt ved problemstillingen. Reguleringsmyndigheter har begrenset informasjon og det er kostbart å inndrive. Aktører, spesielt de som bryter loven, vil ha incentiv til å feilrapportere og holde ting skjult fra myndigheter. Kan forsøke å bruke data på smart måte til å få informasjon..
